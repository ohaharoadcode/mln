## 常用激活函数

### 1.sigmoid函数

$$ f(x)=\frac{1}{1+e^{-x}}$$

其导数为：$$\frac{df(x)}{dx}=\frac{e^{-x}}{(1+e^{-x})^2}=\frac{1}{1+e^{-x}}(1-\frac{1}{1+e^{-x}})=f(x)\times(1-f(x))$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def adjust_ax(ax):
    ax.spines['right'].set_color('none')  #将右侧轴颜色设为无色
    ax.spines['top'].set_color('none')   #上侧轴颜色设为无色
    ax.xaxis.set_ticks_position('bottom') #设置x轴显示的坐标点在下轴上
    ax.spines['bottom'].set_position(('data',0))  
    ax.yaxis.set_ticks_position('left')#设置y轴显示的坐标点在左轴上
    ax.spines['left'].set_position(('data',0))
```

```python
x=np.linspace(-10,10,100)
y=1/(1+np.exp(-x))
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(121)
ax.plot(x,y)
adjust_ax(ax)
t=plt.title('sigmoid',fontsize=20)
ax2=fig.add_subplot(122)
y2=y*(1-y)
ax2.plot(x,y2)
adjust_ax(ax2)
t=plt.title('derivative of sigmoid',fontsize=20)
```

#### 不足：
    在深度神经网络中梯度反向传递时导致梯度爆炸和梯度消失，其中梯度爆炸发生的概率非常小，而梯度消失发生的概率比较大。
    Sigmoid 的 output 不是0均值（即zero-centered）。这是不可取的，因为这会导致后一层的神经元将得到上一层输出的非0均值的信号作为输入。 

### 2.tanh函数

$$tanh=\frac{e^x-e^{-x}}{e^x+e^{-x}}$$

其导数为：$$\frac{d\ tanh}{d x}=1-\frac{(e^x-e^{-x})^2}{(e^x+e^{-x})^2}=\frac{4}{(e^x+e^{-x})^2}$$

```python
x=np.linspace(-10,10,100)
ex=np.exp(x)
y=(ex-1/ex)/(ex+1/ex)
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(121)
ax.plot(x,y)
adjust_ax(ax)
t=plt.title('tanh',fontsize=20)
ax2=fig.add_subplot(122)
y2=4/(ex+1/ex)**2
ax2.plot(x,y2)
adjust_ax(ax2)
t=plt.title('derivative of tanh',fontsize=20)
```

解决了Sigmoid函数的不是zero-centered输出问题，然而，梯度消失（gradient vanishing）的问题和幂运算的问题仍然存在。

### 3.Relu函数

$$Relu=max(0,x)$$

其导数为$$\frac{d\ Relu}{dx}=\begin{cases}
0,(x<0)\\
1,(x>0)\\
\end{cases}$$

```python
x=np.linspace(-10,10,101)
x=np.insert(x,np.where(x==0)[0],0)
y=np.zeros_like(x)
y[x>0]=x[x>0]
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(121)
ax.plot(x,y)
ax.set_title('Relu',fontsize=20)
adjust_ax(ax)
y[y>0]=1
y[np.where(y==0)[0][-1]]=1
ax2=fig.add_subplot(122)
ax2.plot(x,y)
ax2.set_title('derivative of Relu',fontsize=20)
adjust_ax(ax2)
```

最常用的激活函数。
#### 优点：
1） 解决了gradient vanishing问题 (在正区间)

2）计算速度非常快，只需要判断输入是否大于0

3）收敛速度远快于sigmoid和tanh

#### 特别注意的问题：
1）ReLU的输出不是zero-centered

2）Dead ReLU Problem，指的是某些神经元可能永远不会被激活，导致相应的参数永远不能被更新。

有两个主要原因可能导致这种情况产生: 

(1) 非常不幸的参数初始化，这种情况比较少见 

(2) learning rate太高导致在训练过程中参数更新太大，不幸使网络进入这种状态。

解决方法是可以采用Xavier初始化方法，以及避免将learning rate设置太大或使用adagrad等自动调节learning rate的算法。


### 4.Leaky ReLU函数（PReLU）

$$f(x)=max(αx,x)(0<\alpha<1)$$

其导数为$$\frac{df}{dx}=\begin{cases}
\alpha,\ (x<0)\\
\\
1,\ (x>0)\\
\end{cases}$$

```python
alpha=0.1
x=np.linspace(-10,10,101)
x=np.insert(x,np.where(x==0)[0],0)
y=np.zeros_like(x)
y[x>0]=x[x>0]
y[x<0]=alpha*x[x<0]
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(121)
ax.plot(x,y)
ax.set_title('Leaky Relu',fontsize=20)
adjust_ax(ax)
y[y>0]=1
y[y<0]=alpha
index=np.where(y==0)[0]
y[index[-1]]=1
y[index[0]]=alpha
ax2=fig.add_subplot(122)
ax2.plot(x,y)
ax2.set_title('derivative of Relu',fontsize=20)
adjust_ax(ax2)
```

理论上来讲，Leaky ReLU有ReLU的所有优点，外加不会有Dead ReLU问题，但是在实际操作当中，并没有完全证明Leaky ReLU总是好于ReLU。

### 5.ELU (Exponential Linear Units) 函数

$$f(x)=\begin{cases}
x,\ (x\ge0)\\
\ \\
\alpha(e^x-1),\ (x<0)\\
\end{cases}$$

其导数为：$$\frac{df}{dx}=\begin{cases}
1,\ (x>0)\\
\\
\alpha{e^x},\ (x<0)\\
\end{cases}$$

```python
alpha=1
x=np.linspace(-10,10,101)
x=np.insert(x,np.where(x==0)[0],0)
ex=np.exp(x[x<0])
y=np.zeros_like(x)
y[x>0]=x[x>0]
y[x<0]=alpha*(ex-1)
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(121)
ax.plot(x,y)
ax.set_title('Leaky ELU',fontsize=20)
adjust_ax(ax)
y[y>=0]=1
y[y<0]=alpha*ex
# index=np.where(y==0)[0]
# y[index[-1]]=1
y[index[0]]=alpha
ax2=fig.add_subplot(122)
ax2.plot(x,y)
ax2.set_title('derivative of ELU',fontsize=20)
adjust_ax(ax2)
```
